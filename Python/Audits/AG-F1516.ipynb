```python
import pandas as pd
import json
import codecs
import re
import numpy as np
import os
import re
import string
```

```python
files = os.listdir('./FY1516')
text_list = []
name_list = []
for file in files:
    if re.search('.txt', file):
        with codecs.open('./FY1516/' + file,  encoding='utf-8', errors='ignore') as f:
            text = f.read()
            text_list.append(text)
            name_list.append(file)
```

```python
names = [re.sub('cdf.pdf.txt|[0-9]|\.|- ','', name) for name in name_list]
names = [name.lstrip() for name in names]
audits = pd.DataFrame({'file' : names, 'text': text_list})
len(text_list)
```

```python
names
```

```python
### No mention of Unqualified Opinion

print('---opinion---')
### Unqualified Opinion ### 
audits['Unqualified'] = pd.Series(audits.text.str.contains(r'Unqualified Opinion'))
print("Unqualified:", sum(audits.Unqualified))

### Qualified Opinion ### 
audits['Qualified'] = pd.Series(audits.text.str.contains(r'Qualified Opinion'))
print("Qualified:", sum(audits.Qualified))

### Adverse Opinion ### 
audits['Adverse'] = pd.Series(audits.text.str.contains(r'Adverse Opinion'))
print("Adverse:", sum(audits.Adverse))

### Disclaimer of Opinion ###
audits['Disclaimer'] = pd.Series(audits.text.str.contains(r'Disclaimer of Opinion'))
print("Disclaimer:", sum(audits.Disclaimer))

print('----------')
print('total audits:', len(audits))
print("Qualified + Adverse + Disclaimer = ", sum(audits.Qualified) + sum(audits.Adverse) + sum(audits.Disclaimer))

### Check Missing ###
audits['No Opinion'] = ~audits['Qualified'] & ~audits['Adverse'] & ~audits['Disclaimer']
missing = audits[audits['No Opinion']==True]
print('no opinion stated:', sum(audits['No Opinion']))


print('---types---')
### Busary ###
audits['Bursary'] = pd.Series(audits.text.str.contains(r'(?i)bursary|(?i)bursaries'))
print("Bursary:", sum(audits.Bursary))

### Contract ###
audits['Contract'] = pd.Series(audits.text.str.contains(r'(?i)contract|(?i)tender'))
print("Contract:", sum(audits.Contract))

### Schools ###
audits['School'] = pd.Series(audits.text.str.contains(r'(?i)school'))
print("School:", sum(audits.School))

### Students ###
audits['Student'] = pd.Series(audits.text.str.contains(r'(?i)student'))
print("Student:", sum(audits.Student))

### Water ###
audits['Water'] = pd.Series(audits.text.str.contains(r'(?i)water'))
print("Water:", sum(audits.Water))

### Health ###
audits['Health'] = pd.Series(audits.text.str.contains(r'(?i)medical|(?i)health'))
print("Health:", sum(audits.Health))

### Emergency ###
audits['Emergency'] = pd.Series(audits.text.str.contains(r'(?i)emergency'))
print("Emergency:", sum(audits.Emergency))

print('---mismanagement---')
### Late ###
audits['Late'] = pd.Series(audits.text.str.contains(r'(?i)late'))
print("Late:", sum(audits.Late))

### Underutilization ###
audits['Underutilized'] = pd.Series(audits.text.str.contains(r'(?i)under expenditure'))
print("Underutilized:", sum(audits.Underutilized))

### Irregular ###
audits['Irregular'] = pd.Series(audits.text.str.contains(r'(?i)irregular|(?i)irregularities'))
print("Irregular:", sum(audits.Irregular))

### Unaccounted ###
audits['Unaccounted'] = pd.Series(audits.text.str.contains(r'(?i)Un-(?i)Accounted|(?i)Unaccounted'))
print("Unaccounted:", sum(audits.Unaccounted))
```

```python
## Create list of amounts
amountList = pd.DataFrame(columns=['amount', 'file'])
for i in range(0,len(amounts)): 
    temp = pd.DataFrame({'amount': amounts[i], 'file' : audits.file[i]})
    amountList = amountList.append(temp)
amountList[0:10]
```

```python
audit_incidents = audits.merge(amountList, on='file', how='left')
audit_incidents[0:10]

amountList.to_csv('amountList.csv', index=False)
```
